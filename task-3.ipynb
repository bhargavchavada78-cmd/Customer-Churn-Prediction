**import Required libraries,import machine learning model,import numerical operations**

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.linear_model import LinearRegression
from sklearn.ensemble  import RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import confusion_matrix
from sklearn.metrics import classification_report
from sklearn.metrics import accuracy_score,confusion_matrix
from sklearn.preprocessing import StandardScaler


```

**Load the datasets**

```python
df = pd.read_csv('Churn_Modelling.csv')
```

```python
df.info()
```

**Drop non-important Colums**

```python
drop_colm = ['RowNumber','CustomerId','Surname',]
df.drop(drop_colm,axis=1,inplace=True)
```

**Convert categorical variable into numerical form using one-hot encoding**

```python
df = pd.get_dummies(df,drop_first=True)
```

**'Exited' is the target column**

```python
x = df.drop('Exited',axis=1)
y = df['Exited']
```

**Align training and testing datasets to ensure same feature columns**



```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=42)
```

**Scale the feature values to improve model performance**

```python
Scaler = StandardScaler()
x_train = Scaler.fit_transform(x_train)
x_test = Scaler.transform(x_test)
```

```python
print(x_train.shape)
print(y_train.shape)
print(y_train.isna().sum())
```

**Initialize the LogisticRegression**





```python
lr = LogisticRegression()
lr.fit(x_train,y_train)
```

**Predict fraud on test data**

```python
y_pred = lr.predict(x_test)
```

**Evaluate model performance**

```python
print(classification_report(y_test,y_pred))
```

**Initialize the RandomForestClassifier**

```python
rn = RandomForestClassifier()
rn.fit(x_train,y_train)
```

**Predict fraud on test data**

```python
y_pred=rn.predict(x_test)
```

**Evaluate model performance**

```python
print(classification_report(y_test,y_pred))
```

**Intialize the Decision Tree Classifier**

```python
tr = DecisionTreeClassifier()
tr.fit(x_train,y_train)
```

**Predict fraud on test data**

```python
y_pred=tr.predict(x_test)
```

**Evaluate model performance**

```python
print(classification_report(y_test,y_pred))
```

**Take input from user**

```python
new_customer = [[600,40,5,60000,2,1,1,50000,False,True,True]]
```

```python
new_customer = Scaler.transform(new_customer)



```

**Predict Final Result**



```python
prediction = rn.predict(new_customer)

if prediction == 1:
    print("⚠️Customer will churn")
else:
    print("✅Customer will stay")
```
